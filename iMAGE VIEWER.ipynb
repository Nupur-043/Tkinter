```python
from PIL import ImageTk, Image
root = Tk()
root.title('cn project learning')

iconpath=r"C:\Users\KIIT\Desktop\5th sem materials\Project\python_squre_icon_icon_257067.ico"
root.iconbitmap(iconpath)


my_img1 =  ImageTk.PhotoImage(Image.open('images/dog.png'))
my_img2 =  ImageTk.PhotoImage(Image.open('images/crack.png'))
my_img3 =  ImageTk.PhotoImage(Image.open('images/earth.png'))
my_img4 =  ImageTk.PhotoImage(Image.open('images/butterfly.png'))
my_img5 =  ImageTk.PhotoImage(Image.open('images/tiger.png'))
my_img6 =  ImageTk.PhotoImage(Image.open('images/scenery.png'))
image_list = [my_img1, my_img2, my_img3, my_img4, my_img5, my_img6]

my_label = Label(image=my_img1)
my_label.grid(row=0, column=0 , columnspan=3)


def fwd(imgno):
    global my_label
    global button_fwd
    global button_back
   
    my_label.grid_forget()
    my_label= Label(image=image_list[imgno-1])
    button_fwd = Button(root, text='>>', command=lambda: fwd(imgno+1))
    button_back = Button(root, text='<<', command=lambda: back(imgno-1))
    
    if imgno==5:
        button_fwd=Button(root, text='>>',state=DISABLED)
    
    my_label.grid(row=0, column=0 , columnspan=3)
    button_back.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    
    
def back(imgno):
    global my_label
    global button_fwd
    global button_back
    
    my_label.grid_forget()
    my_label= Label(image=image_list[imgno-1])
    button_fwd = Button(root, text='>>', command=lambda: fwd(imgno+1))
    button_back = Button(root, text='<<', command=lambda: back(imgno-1))
    
    if imgno==1:
        button_back=Button(root, text='>>',state=DISABLED)
    
    my_label.grid(row=0, column=0 , columnspan=3)
    button_back.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    

button_back = Button(root, text='<<', command=back, state=DISABLED)
button_exit = Button(root, text='Exit program', command=root.quit)
button_fwd = Button(root, text='>>', command=lambda: fwd(2))



button_back.grid(row=1, column=0)
button_exit.grid(row=1, column=1)
button_fwd.grid(row=1, column=2)

#my_but = ImageTk.PhotoImage(Image.open('vid.png'))
#button_vid = Button(root, te   xt='click to turn on the camera',image=my_but, padx=40, pady=20)
#button_vid.grid(row=3, column=0)










#button_quit = Button(root, text="exit page", command=root.quit)
#button_quit.pack()


root.mainloop()
```

## ADding a status bar

```python
from PIL import ImageTk, Image
root = Tk()
root.title('cn project learning')

iconpath=r"C:\Users\KIIT\Desktop\5th sem materials\Project\python_squre_icon_icon_257067.ico"
root.iconbitmap(iconpath)


my_img1 =  ImageTk.PhotoImage(Image.open('images/dog.png'))
my_img2 =  ImageTk.PhotoImage(Image.open('images/crack.png'))
my_img3 =  ImageTk.PhotoImage(Image.open('images/earth.png'))
my_img4 =  ImageTk.PhotoImage(Image.open('images/butterfly.png'))
my_img5 =  ImageTk.PhotoImage(Image.open('images/tiger.png'))
my_img6 =  ImageTk.PhotoImage(Image.open('images/scenery.png'))

image_list = [my_img1, my_img2, my_img3, my_img4, my_img5, my_img6]

#define
status = Label(root, text="Image 1 of " + str(len(image_list)), bd=1, relief=SUNKEN,
              anchor=W)

my_label = Label(image=my_img1)
my_label.grid(row=0, column=0 , columnspan=3)


def fwd(imgno):
    global my_label
    global button_fwd
    global button_back
   
    my_label.grid_forget()
    my_label= Label(image=image_list[imgno-1])
    button_fwd = Button(root, text='>>', command=lambda: fwd(imgno+1))
    button_back = Button(root, text='<<', command=lambda: back(imgno-1))
    
    if imgno==5:
        button_fwd=Button(root, text='>>',state=DISABLED)
    
    my_label.grid(row=0, column=0 , columnspan=3)
    button_back.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    
    # Update status bar
    status = Label(root, text="Image"+f"{imgno}"+" of " + str(len(image_list)), bd=1, relief=SUNKEN,
              anchor=W)
    status.grid(row=2, columnspan=3, sticky=W+E)
    
    
def back(imgno):
    global my_label
    global button_fwd
    global button_back
    
    my_label.grid_forget()
    my_label= Label(image=image_list[imgno-1])
    button_fwd = Button(root, text='>>', command=lambda: fwd(imgno+1))
    button_back = Button(root, text='<<', command=lambda: back(imgno-1))
    
    if imgno==1:
        button_back=Button(root, text='>>',state=DISABLED)
    
    my_label.grid(row=0, column=0 , columnspan=3)
    button_back.grid(row=1, column=0)
    button_fwd.grid(row=1, column=2)
    

button_back = Button(root, text='<<', command=back, state=DISABLED)
button_exit = Button(root, text='Exit program', command=root.quit)
button_fwd = Button(root, text='>>', command=lambda: fwd(2))



button_back.grid(row=1, column=0)
button_exit.grid(row=1, column=1)
button_fwd.grid(row=1, column=2,pady=10)
status.grid(row=2, columnspan=3, sticky=W+E) 
#sticky east west north south if we have to stretch from west to east



root.mainloop()
```
